```python
import matplotlib.pyplot as plt
word_count = {}
with open('hw2_data.txt', 'r')as file:
    for line in file:
        word = line.strip().lower() #去除換行並轉小寫
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

# 1. 印出不重複的英文字數
unique_words = len(word_count)
print(f"不重複的英文字總數:{unique_words}")

# 2. 印出每個英文字的出現次數
print("\n每個英文字出現次數：")
for word, count in sorted(word_count.items(), key=lambda x: x[1], reverse=True):
    print(f"{word}: {count}")

# 3. 畫直方圖
words_sorted = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
words = [item[0] for item in words_sorted]
counts = [item[1] for item in words_sorted]

plt.figure(figsize=(18, 8), dpi=100)
bars = plt.bar(words[:20], counts[:20], color='#FFBB77')  # 只畫前 20 個單字

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, yval, 
             ha='center', va='bottom', fontsize=10)


plt.title('Word Frequency Histogram', fontsize=20, fontweight='bold')
plt.xlabel('Word', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.xticks(rotation=0, ha='center', fontsize=14) # 橫向顯示 x 軸文字 # ha=horizontalalignment
plt.yticks(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('fancy_histogram.png')
plt.show()
```
